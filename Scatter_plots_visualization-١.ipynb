```python

import pandas as pd # used to wirk with excel sheet files
import numpy as np

#data visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('data.csv')
print(data.shape)
data.head(10) #display the firsy 10 rows
```

# data explaining :

* the data is consists of nearly 3800 farmer from india
* each farmer has a unique id (ID column).
* the (district) column show the farmer district.
* the Acre shows the area of cultivated land in acre.
* the yield shows the yield of cultivation per acre

# ❗ we want to study the correlation between the acre and yield❗

# using seaborn library :

* u can easily plot the scatterplot using seaborn scatterplot() method :

```python
sns.scatterplot(data,x='Acre',y='Yield')
```

## also you can use regplot() method wich plot the scatter and a line wich fit the scatter

```python
sns.regplot(data,x='Acre',y='Yield')
```

# using matplotlib library :

* you can rach the same goals using matplotlib

```python
plt.scatter(data.Acre,data.Yield)
```

# Conclusion :

* as we saw, the scatter plots gives us a sense (or sometime confidence about what the relation between two variables .

* a small tip :
we can also calculate the correlation using this line of code :

```python
data[['Acre','Yield']].corr()
```
